```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
import os

# Create data folder (Colab doesn't have it by default)
os.makedirs("data", exist_ok=True)

# ---------------------------
# Configuration
# ---------------------------
num_days = 60
transactions_per_day = (50, 120)
start_date = datetime(2025, 1, 1)

menu = [
    ("Margarita", "Cocktail", 12),
    ("Old Fashioned", "Cocktail", 14),
    ("Mojito", "Cocktail", 11),
    ("Vodka Soda", "Cocktail", 10),
    ("Whiskey Sour", "Cocktail", 13),
    ("IPA Beer", "Beer", 8),
    ("Lager Beer", "Beer", 7),
    ("Stout Beer", "Beer", 9),
    ("Red Wine", "Wine", 10),
    ("White Wine", "Wine", 9),
    ("Tequila Shot", "Shot", 6),
    ("Vodka Shot", "Shot", 6),
    ("Rum & Coke", "Cocktail", 11)
]

payment_types = ["Card", "Cash"]

data = []

def generate_time():
    hour_weights = (
        [1]*6 +
        [2]*6 +
        [4]*5 +
        [8]*7
    )
    hours = list(range(24))
    hour = random.choices(hours, weights=hour_weights, k=1)[0]

    minute = random.randint(0, 59)
    second = random.randint(0, 59)

    return f"{hour:02d}:{minute:02d}:{second:02d}"

# ---------------------------
# Generate dataset
# ---------------------------
for day in range(num_days):
    current_date = start_date + timedelta(days=day)

    if current_date.weekday() >= 4:  # Fri-Sun
        num_transactions = random.randint(80, 150)
    else:
        num_transactions = random.randint(*transactions_per_day)

    for _ in range(num_transactions):
        item_name, category, base_price = random.choice(menu)

        price = round(base_price + random.uniform(-1, 1), 2)
        quantity = np.random.choice([1, 1, 1, 2, 2, 3])
        payment = random.choices(payment_types, weights=[0.8, 0.2])[0]

        time_str = generate_time()

        data.append([
            current_date.strftime("%Y-%m-%d"),
            time_str,
            item_name,
            category,
            quantity,
            price,
            payment
        ])

# Create DataFrame
df = pd.DataFrame(data, columns=[
    "date", "time", "item_name", "category",
    "quantity", "price", "payment_type"
])

# Save file
file_path = "data/bar_sales.csv"
df.to_csv(file_path, index=False)

print("CSV created at:", file_path)
print("Rows:", len(df))

df.head()
```
